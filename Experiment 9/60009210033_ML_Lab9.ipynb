```python
#TASK 1
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import warnings
# Set the random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
n_samples = 200
n_clusters = 3
mean_value = 2.7

# Generate data with 3 clusters
X = np.concatenate((np.random.normal(mean_value, 0.5, (n_samples // 3, 2)),
                    np.random.normal(mean_value + 3, 0.5, (n_samples // 3, 2)),
                    np.random.normal(mean_value - 3, 0.5, (n_samples // 3, 2))))

# Perform K-means clustering
lowest_sse = np.inf
final_centroids = None
num_iterations = None
predicted_labels = None

# Perform K-means clustering with 10 variations of initial means
for i in range(10):
    kmeans = KMeans(n_clusters=n_clusters, init='random', max_iter=300, random_state=i)
    kmeans.fit(X)

    # Get SSE value
    sse = kmeans.inertia_
    if sse < lowest_sse:
        lowest_sse = sse
        final_centroids = kmeans.cluster_centers_
        num_iterations = kmeans.n_iter_
        predicted_labels = kmeans.labels_

# Plot the data points and centroids
plt.scatter(X[:, 0], X[:, 1], c=predicted_labels, cmap='viridis')
plt.scatter(final_centroids[:, 0], final_centroids[:, 1], c='red', marker='X')
plt.title("K-means Clustering")
plt.xlabel("X")
plt.ylabel("Y")
plt.figure(figsize=(10, 10))
plt.show()
```

```python
# TASK 2
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
n_samples = 200
mean_value = 2.7

# Generate data with 3 clusters
X = np.concatenate((np.random.normal(mean_value, 0.5, (n_samples // 3, 2)),
                    np.random.normal(mean_value + 3, 0.5, (n_samples // 3, 2)),
                    np.random.normal(mean_value - 3, 0.5, (n_samples // 3, 2))))

# Calculate SSE for different number of clusters
sse_values = []
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, init='random', max_iter=300, random_state=0)
    kmeans.fit(X)
    sse = kmeans.inertia_
    sse_values.append(sse)

# Plot SSE values against number of clusters (elbow method)
plt.plot(range(2, 11), sse_values, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()

# Calculate silhouette scores for different number of clusters
silhouette_scores = []
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, init='random', max_iter=300, random_state=0)
    kmeans.fit(X)
    labels = kmeans.labels_
    silhouette_scores.append(silhouette_score(X, labels))

# Plot silhouette scores against number of clusters (silhouette method)
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title("Silhouette Method")
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.show()
```

```python
#TASK 3
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic_df = pd.read_csv('/content/train(1).csv')

# Preprocessing
# Drop unnecessary columns
titanic_df = titanic_df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked'], axis=1)

# Convert categorical variables to numerical
titanic_df['Sex'] = titanic_df['Sex'].map({'male': 0, 'female': 1})

# Handling missing values
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())

# Standardize the features
scaler = StandardScaler()
X = scaler.fit_transform(titanic_df)

# Perform k-means clustering
k = 2  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
labels = kmeans.fit_predict(X)

# Evaluate clustering results
silhouette_avg = silhouette_score(X, labels)
print("Silhouette Score:", silhouette_avg)

# Plotting the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', color='red', label='Centroids')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-means Clustering')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Step 1: Load the Titanic training dataset
titanic_data = pd.read_csv('/content/train(1).csv')

# Step 2: Preprocess the data
# Drop irrelevant columns and handle missing values
titanic_data = titanic_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked'], axis=1)
titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace=True)

# Convert categorical variables to numerical using one-hot encoding
titanic_data = pd.get_dummies(titanic_data, columns=['Sex', 'Pclass'])

# Step 3: Perform feature scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(titanic_data)

# Step 4: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(scaled_data, titanic_data['Survived'], test_size=0.2, random_state=42)

# Step 5: Apply K-means clustering
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_train)

# Step 6: Predict the clusters for the training and testing data
train_clusters = kmeans.predict(X_train)
test_clusters = kmeans.predict(X_test)

# Step 7: Calculate the accuracy of K-means clustering on the testing data
accuracy = accuracy_score(y_test, test_clusters)
print("Accuracy:", accuracy)

# Step 8: Plot the clusters
plt.scatter(X_train[:, 0], X_train[:, 1], c=train_clusters, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', color='red', label='Centroids')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-means Clustering')
plt.legend()
plt.show()
```
